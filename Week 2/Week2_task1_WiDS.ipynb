```python
img1 = cv2.imread("image1.jpg")
img2 = cv2.imread("image2.jpg")

img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

plt.imshow(img1)
plt.title("Image 1")
plt.axis("off")
```

```python
plt.imshow(img2)
plt.title("Image 2")
plt.axis("off")
```

```python
gray1 = cv2.cvtColor(img1, cv2.COLOR_RGB2GRAY)
gray2 = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)

plt.imshow(gray1, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")
```

```python
plt.imshow(gray2, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")
```

```python
plt.hist(gray1.ravel(), bins=256)
plt.title("Grayscale Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(gray2.ravel(), bins=256)
plt.title("Grayscale Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.show()
```

```python
resized1 = cv2.resize(img1, (256, 256))
resized2 = cv2.resize(img2, (256, 256))
plt.imshow(resized1)
plt.title("Resized Image")
plt.axis("off")
```

```python
plt.imshow(resized2)
plt.title("Resized Image")
plt.axis("off")
```

```python
edges1 = cv2.Canny(gray1, 100, 200)
edges2 = cv2.Canny(gray2, 100, 200)
plt.imshow(edges1, cmap="gray")
plt.title("Edge Detection")
plt.axis("off")
```

```python
plt.imshow(edges2, cmap="gray")
plt.title("Edge Detection")
plt.axis("off")
```

```python
noise = np.random.normal(0, 25, gray1.shape)
noisy_image1 = gray1 + noise
noisy_image1 = np.clip(noisy_image1, 0, 255).astype(np.uint8)

plt.imshow(noisy_image1, cmap="gray")
plt.title("Noisy Image 1")
plt.axis("off")
```

```python
noise = np.random.normal(0, 25, gray2.shape)
noisy_image2 = gray2 + noise
noisy_image2 = np.clip(noisy_image2, 0, 255).astype(np.uint8)

plt.imshow(noisy_image2, cmap="gray")
plt.title("Noisy Image 2")
plt.axis("off")
```

```python
denoised = cv2.GaussianBlur(noisy_image1, (5, 5), 0)

plt.imshow(denoised, cmap="gray")
plt.title("Denoised Image")
plt.axis("off")
```

```python
denoised = cv2.GaussianBlur(noisy_image2, (5, 5), 0)

plt.imshow(denoised, cmap="gray")
plt.title("Denoised Image")
plt.axis("off")
```
