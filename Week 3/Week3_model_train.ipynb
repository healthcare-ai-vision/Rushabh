```python
from ultralytics import YOLO
from roboflow import Roboflow
import os
```

```python
rf = Roboflow(api_key="2hHNdIQGbzDmN5Uw76QJ")

project = rf.workspace("cxrdataset").project("skin-ga5ww")
dataset = project.version(1).download("folder")
```

```python
model = YOLO("yolov8n-cls.pt")   # Pretrained ImageNet classifier
```

```python
model.train(
    data=dataset.location,
    epochs=25,
    imgsz=224,
    batch=32,
    device=0
)
```

```python
metrics = model.val()

print("Top-1 Accuracy:", metrics.top1)
```
